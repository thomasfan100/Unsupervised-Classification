## Reproduction Methodology

For code reproduction, my group had to make some small modifications when following the instructions provided by the author. After downloading the code provided, the env.yml file must be renamed and updated based on the local machine path and the file paths to the direct datasets used must also be changed as well. Since our code was run using the resources provided on the Mill, our job had to install dependencies a little differently, especially given that it now 2024 and the versions that were installed for the SCAN paper are now deprecated. Here is what we did to run it on Missouri S&T's newest high-performance cluster. First, we loaded necessary modules including anaconda and cuda-toolkit. Then, we load the virtual conda environment to install the important packages. We used conda to install the deprecated pytorch, and other libraries but pip was needed to download the torchvision package.

The training was accomplished by using the batch scripts with the commands provided by the paper. You can change the running epochs easily in the config files depending on which task you want to run and the data is saved into memory so a batch can be cancelled and run again afterwards. Our batch script runs for 1 task and allocates 1 GPU node and the results are outputted in a out file afterwards.

## Challenges Encountered

A brief rundown of some challenges encountered with the source code and environment will be detailed. I will first briefly share some challenges we encountered when trying to run the tutorial code provided. Since this code uses a deprecated version of pytorch, it may cause some issues with your installed version. For example, I realized that collections.Mapping and collections.Sequence have changed since then. Furthremore, the tutorial code mentions the need to add some repository_eccv folder but they didnt mention that the config file must be changed as well so the base directory path can look into there in the future. Another small thing to mention is that the pretrained models need to be renamed to checkpoint.pth.tar as well. After we were able to reproduce the tutorial, we moved on to the actual implementation of all three steps described in SCAN. In this phase, we had issues requesting a GPU on the Mill and saw many issues regarding cuda not being present. The easiest way to resolve this is to simply submit the jobs as scripts rather then attempting to use an interactive session. Furthermore, we had some issues with the source code as well. One thing to note is that the code will download the datasets for you and you can simply specify the path in the configs. We are relatively new with tools like the GPU and Google Colab and so some of these seemingly simple setup issues were a slight bottleneck to us but since the original paper is really well documented, our group only faced problems mostly on environment setup and custom unit tests.
